```python
import pandas as pd
import numpy as np
import io
import requests
```

```python
#Primero obtenemos los datasets con la libreria requests y pandas.
#Dataset amazon prime titles

url_amzn = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/amazon_prime_titles.csv'
bajada_datos_amzn = requests. get(url_amzn). content
amzn_df = pd. read_csv(io. StringIO(bajada_datos_amzn. decode('utf-8')))

#Dataset disney plus titles

url_disney = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/disney_plus_titles.csv'
bajada_datos_disney = requests. get(url_disney). content
disney_df = pd. read_csv(io. StringIO(bajada_datos_disney. decode('utf-8')))


#Dataset hulu titles

url_hulu = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/hulu_titles.csv'
bajada_datos_hulu = requests. get(url_hulu). content
hulu_df = pd. read_csv(io. StringIO(bajada_datos_hulu. decode('utf-8')))


#Dataset netflix titles

url_netflix = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/netflix_titles.json'
bajada_datos_netflix = requests. get(url_netflix). content
netflix_df = pd. read_json(io. StringIO(bajada_datos_netflix. decode('utf-8')))

```

```python
# 1. EDA para los datasets.

#eliminamos la columna 'show_id' ya que es un index repetido
amzn_df.drop('show_id', axis=1, inplace=True)
amzn_df.drop('description', axis=1, inplace=True)

disney_df.drop('show_id', axis=1, inplace=True)
disney_df.drop('description', axis=1, inplace=True)

hulu_df.drop('show_id', axis=1, inplace=True)
hulu_df.drop('description', axis=1, inplace=True)

netflix_df.drop('show_id', axis=1, inplace=True)
netflix_df.drop('description', axis=1, inplace=True)

# Normalizacion de columnas
amzn_df.columns = ['Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_añadida', 'Año_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en']
disney_df.columns = ['Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_añadida', 'Año_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en']
hulu_df.columns = ['Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_añadida', 'Año_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en']
netflix_df.columns = ['Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_añadida', 'Año_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en']
```

```python
# Valores faltantes para Amazon.

total = amzn_df.isnull().sum().sort_values(ascending=False)
porcentaje = (amzn_df.isnull().sum()/amzn_df.isnull().count()).sort_values(ascending=False)
valores_faltantes = pd.concat ([total, porcentaje], axis=1, keys=['Total', 'Porcentaje'])
valores_faltantes.head(11)
```

```python
# Añadimos una columna llamada 'Plataforma' a los data frames.

lista_1=['Amazon']*8964
amzn_df['Plataforma']=lista_1

lista_2=['Disney']*1389
disney_df['Plataforma']= lista_2

lista_3=['Hulu']*2814
hulu_df['Plataforma']= lista_3

lista_4=['Netflix']*8265
netflix_df['Plataforma']= lista_4
```

```python
# Reindexamos los dfs.

amzn_df=amzn_df.reset_index()
disney_df=disney_df.reset_index()
hulu_df=hulu_df.reset_index()
netflix_df=netflix_df.reset_index()
```

```python
# Normalizamos las mayusculas y minúsculas de algunas columnas.

# Para ello, creamos las siguiente funcion.

def funcion_title_data (text):
    if type(text) != str:
        return 'Error'
    else:
        text_normal = text.title()
        return (text_normal)

    
# Aplicamos la funcion a las columnas de los 4 datasets:

amzn_df['Tipo']=amzn_df['Tipo'].apply(funcion_title_data)
amzn_df['Titulo']=amzn_df['Titulo'].apply(funcion_title_data)
amzn_df['Director']=amzn_df['Director'].apply(funcion_title_data)
amzn_df['Elenco']=amzn_df['Elenco'].apply(funcion_title_data)
amzn_df['Listada_en']=amzn_df['Listada_en'].apply(funcion_title_data)
amzn_df['Duracion']=amzn_df['Duracion'].apply(funcion_title_data)

disney_df['Tipo']=disney_df['Tipo'].apply(funcion_title_data)
disney_df['Titulo']=disney_df['Titulo'].apply(funcion_title_data)
disney_df['Director']=disney_df['Director'].apply(funcion_title_data)
disney_df['Elenco']=disney_df['Elenco'].apply(funcion_title_data)
disney_df['Listada_en']=disney_df['Listada_en'].apply(funcion_title_data)
disney_df['Duracion']=disney_df['Duracion'].apply(funcion_title_data)

hulu_df['Tipo']=hulu_df['Tipo'].apply(funcion_title_data)
hulu_df['Titulo']=hulu_df['Titulo'].apply(funcion_title_data)
hulu_df['Director']=hulu_df['Director'].apply(funcion_title_data)
hulu_df['Elenco']=hulu_df['Elenco'].apply(funcion_title_data)
hulu_df['Listada_en']=hulu_df['Listada_en'].apply(funcion_title_data)
hulu_df['Duracion']=hulu_df['Duracion'].apply(funcion_title_data)

netflix_df['Tipo']=netflix_df['Tipo'].apply(funcion_title_data)
netflix_df['Titulo']=netflix_df['Titulo'].apply(funcion_title_data)
netflix_df['Director']=netflix_df['Director'].apply(funcion_title_data)
netflix_df['Elenco']=netflix_df['Elenco'].apply(funcion_title_data)
netflix_df['Listada_en']=netflix_df['Listada_en'].apply(funcion_title_data)
netflix_df['Duracion']=netflix_df['Duracion'].apply(funcion_title_data)
```

Finalmente concatenamos los 4 datses ya limpios.

```python
concat_1 = pd.concat([amzn_df, disney_df], axis=0)
concat_2=pd.concat([hulu_df,netflix_df], axis=0)
movies_df=pd.concat([concat_1,concat_2], axis=0)
```

```python
#Le hacemos un arreglo al index del nuevo dataset

movies_df.drop('index', inplace=True, axis=1)
movies_df=movies_df.reset_index()
```

 Ahora vamos a realizar unas modificaciones para realizar nuestras consultas.

```python
cat = movies_df['Duracion'].str.split(" ", n = 1, expand = True) 
  
movies_df['Duracion']= cat[0] 
movies_df['unidad_tiempo']=cat[1]
```

```python
movies_df['Duracion']=movies_df['Duracion'].astype(int)
movies_df['unidad_tiempo']=movies_df['unidad_tiempo'].replace('Seasons','Season')
```
